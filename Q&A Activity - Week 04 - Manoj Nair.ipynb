# Qn. 4. Find out which contract is preferred by the senior citizen.

```python
plt.figure(figsize=(15,10))

sns.countplot(x='SeniorCitizen',data=tel_churn,hue="Contract")
plt.title('Senior Citizen Prefered Contract',size=20)
plt.xlabel('Senior Citizen')
plt.ylabel('Contract Count')
plt.show()
```

##### Qn. 4 Insights
* Senior citizens prefers month to month contracts. 
* Opted out senior citizen customers prefer 1 year contracts than 2 year
* Remaining senior citizen customers prefer 2 year contracts than 1 year

# Qn. 5. Comment on your finds on Payment Method

```python
plt.figure(figsize=(15,10))

sns.countplot(x='SeniorCitizen',data=tel_churn,hue="PaymentMethod")
plt.title('Payment Method Based on Citizen Category ',size=20)
plt.xlabel('Senior Citizen')
plt.ylabel('Contract Count')
plt.show()
```

```python
plt.figure(figsize=(15,10))

sns.countplot(x='gender',data=tel_churn,hue="PaymentMethod")
plt.title('Payment method Based on Gender',size=20)
plt.xlabel('Gender')
plt.ylabel('Contract Count')
plt.show()
```

##### Qn. 5 Insights
* Irrespective of customer category, 'Electronic Check' is the most preferred payment method
* For Senior citizen customers; 'Mailed Check' is the least preferred payment method, 'Bank transfer’ & ‘Credit Card’ are almost   equal with a slighter preference for ‘Bank transfer’. 
* Female customers slightly prefer 'Bank transfer' to 'Mailed Check' which is slightly preferred than 
  'Credit Card'
* Male customers prefer 'Mailed check' to 'Credit Card' which is slightly preferred than ‘Bank transfer’
* ‘Credit Card’ is the least preferred in female & ‘Bank transfer’ is the least preferred in the male gender category
