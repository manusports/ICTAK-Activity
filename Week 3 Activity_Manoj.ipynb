#### 4.5, Create a histogram of "Age".

#### 4.6, Create a scatter plot of "BMI" vs "Glucose".

```python
plt.scatter(x = diab['BMI'], y = diab['Glucose'], s = 10)
plt.title('BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.xlim(0, 68)
plt.show()
```

#### 4.7, Replace the min value of "BMI" with mean value of BMI

```python
diab.describe()
```

```python
min_BMI=diab['BMI'].min()
min_BMI
```

```python
mean_BMI=diab['BMI'].mean()
mean_BMI
```

```python
diab['BMI']=diab['BMI'].replace(min_BMI,mean_BMI)
new_BMI_min=diab['BMI'].min()
new_BMI_min
```
